```python
import seaborn as sns
from sklearn.cluster import KMeans

# Select the features for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
# Apply KMeans clustering with 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Plot the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Annual Income (k$)',
    y='Spending Score (1-100)',
    hue='Cluster',
    palette='Set1',
    data=df,
    s=100,
    alpha=0.8
)
plt.title('Customer Segments by Annual Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()
```

<li><b>Cluster 3 - Purple (High Income, Low Spending)</b>

<ul><li>Income: High

<li>Spending: Low

#Basic Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Mall_Customers.csv')
df.head()

```

```python
# 1. Age Distribution Histogram
plt.figure()
plt.hist(df['Age'], bins=10)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.show()
```

Customers range from late teens to around 70 years old, with a concentration between 25–40.There’s a notable tapering in customer counts above age 50.
Hence Marketing campaigns could be tailored primarily toward the 25–40 age bracket, which forms the core customer base.

```python
# 2. Annual Income Distribution Histogram
plt.figure()
plt.hist(df['Annual Income (k$)'], bins=10)
plt.title('Annual Income Distribution')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Number of Customers')
plt.show()
```

Income clusters around the 40k-80k range, with fewer customers at the very low less than 30k and very high greater 100k ends.

Pricing and product positioning can focus on upper-middle-income earners, while special offers might be needed to engage both lower- and higher-income segments.

```python
# 3. Spending Score Distribution Histogram
plt.figure()
plt.hist(df['Spending Score (1-100)'], bins=10)
plt.title('Spending Score Distribution')
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Number of Customers')
plt.show()
```

Spending Scores are fairly evenly spread across the 1-100 range, but there's a slight bulge in the mid-range (around 40-60)

This shows that many customers are moderate spenders. A loyalty or rewards program could encourage mid-range scorers to move into higher-spending tiers.

```python
# 4. Scatter: Annual Income vs Spending Score
plt.figure()
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'])
plt.title('Annual Income vs Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

No strong linear correlation observed. Implies that a one-size-fits-all approach won't work. We will have to implement segmented marketing

```python
# 5. Boxplot: Spending Score by Gender
plt.figure()
df.boxplot(column='Spending Score (1-100)', by='Gender')
plt.title('Spending Score by Gender')
plt.suptitle('')
plt.xlabel('Gender')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

Gender alone isn't a strong differentiator for income or spending
